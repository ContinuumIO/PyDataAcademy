After picking a webpage (in this case a search of cats on youtube) we want to look at both the rendered page and the source page.  Again, we want to store and process each video and releated metada of the video: duration, data uploaded, etc.  Looking at the source code we can see that each video is wrapped in **`<li>`** and **`<div>`** tags.

Before we think about pulling out those tags we first need to grab the web page.  With Python and the [requests](http://docs.python-requests.org/en/latest/) module HTTP requests are trivial.   

```python
import requests, re
r = requests.get('http://www.youtube.com/results?search_query=cats')
```

Using **requests** we download a **Response** object: **r**.  **r** has an intuitive interface for accessing standardized elements of a webpage: content, header, cookies, etc. We don't know or care how the HTML is organized.  After the call to **get** the contents of the page are stored in the _content_ member variable

```python
page = r.content
```

Upon deeper inspection of [source](view-source:http://www.youtube.com/results?search_query=cats), we can see video href link is listed within the **<div>** tag.  For example:

```python
page[16307:16500] #print first 100 characters
```

We can extract all the href tags by writing a [regular expression](http://docs.python.org/2/howto/regex.html) and find all the occurences of each match.
The regular expression below matches text which begins with **`<div class="yt-lockup-thumbnail"><a href="`** and ends with **<code>" class</code>**.  

In the parentheses the **.*** matches any repetition of any character. 

```python
videoLink = re.compile('<div class="yt-lockup-thumbnail"><a href="(.*)" class')
videoLinks = re.findall(videoLink,page)
videoLinks[0]
```

This is a good start but it's not very extensible.  We want to also store the title of the video and number of pages views.  If we continue along this path we'dhave to build multiple and increasingly more complicated regular expressions.  Luckily, there are many Python programmers which are interested in web-scraping and as such, they've built a library: [BeautifulSoup](http://pypi.python.org/pypi/BeautifulSoup)

__BeautifulSoup provides compact methods for searching over tree structures such as HTML and XML.__

```python
from BeautifulSoup import BeautifulSoup
soup = BeautifulSoup(page)

divs = soup('div',{'class':'yt-lockup-content'})

print divs[0]
```

Above give the page we've previously downloaded to be _soupified_.  After _soupification_, we request all the div tags with the class **yt-lockup-content**.  As you can see from the print divs is a list containing all the data in the div.  We can easily extract the links by doing the following

```python
for div in divs:
    print div.a['href'], div.a.text
```

Looking at divs[0] again we see that number of views and other metadata are delievered in span elements.  We can pull those out by calling BeatifulSoup's **findAll** method.

```python
metaData = divs[0].findAll('span')

for i, md in enumerate(metaData):
    print i, md.text
```

Using **enumerate** we quickly see the index and the associated metadata.  Date uploaded is index 2 and views is index 4.  We'd likely to store the  views as ints so we need to do a little more processing.

```python
for div in divs:
    md = div.findAll('span')
   
    dateAdded = (md[2]['class'],md[2].text) #date-added
    
    count = md[4].text #number of page view
    count = count.split(' ')[0] #[u'47,905,925', u'views']

    count = int(count.replace(',','')) #convert viewcounts value to int
    viewcounts = (md[4]['class'],count)

    print dateAdded, viewcounts 
```

This is code for a playlist.  We were focused on videos, not a playlist of videos, which apparently are structured differently.  We can process just the video elements and _not_ the playlists if we are a little more precise.  Instead of starting with `<div>` tags let's instead use the `<li>` tag that we originally inspected.

```python
youtube = []
lis = soup('li',{'class':'yt-grid-box result-item-video context-data-item'})

for li in lis:
    video = {} #empty dictionary
    div = li('div',{'class':'yt-lockup-content'})[0] #get div tag
 
    md = div.findAll('span') #find span tags which contain metadata of video

    video['url'] = div.a['href'] #link
    video['title'] = div.a.text #link name
    video['duration'] = li.a.text #duration of video 

    video[md[2]['class']] = md[2].text #date-added
    
    viewcounts = md[4].text #number of page view
    viewcounts = viewcounts.split(' ')[0] #[u'47,905,925', u'views']

    video[md[4]['class']] = int(viewcounts.replace(',','')) #convert pageviews value to int

    youtube.append(video)
```
